```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
a = soup.find_all('h2',class_="a-size-mini a-spacing-none a-color-base s-line-clamp-3")
```

```python
a
```

Extract product names from scrapped data

```python
product_names = [tag.text.strip() for tag in a]

# Print the extracted product names
for names in product_names:
    print(names)
```

Extract product prices from scrapped data

```python
price = soup.find_all('span',class_="a-price-whole")
```

```python
prices=[tag.text.strip() for tag in price]
```

```python
prices
```

Extract ratings from scrapped data

```python
rating = soup.find_all('i',class_="a-icon-star-small")
```

```python
ratings=[tag.text.strip() for tag in rating]
```

```python
len(ratings)
```

```python
delivery_date = soup.find_all('span',class_="a-color-base a-text-bold")
```

```python
delivery_dates=[tag.text.strip() for tag in delivery_date]
```

```python
delivery_dates
```

Extract number of reviews on each product from scrapped data

```python
review = soup.find_all('span',class_="a-size-base s-underline-text")
```

```python
reviews=[tag.text.strip() for tag in review]
```

```python
reviews
```

Create a data frame

```python
prices.pop()
ratings.pop()
reviews.pop()
len(product_names), len(prices), len(ratings), len(reviews)
#Now we can easily create data frame since all the columns have same number of tuples
```

```python
import numpy as np
import pandas as pd

data = {
    'Product Name': product_names,
    'Price': prices,
    'Rating': ratings,
    'Reviews': reviews,
}

df = pd.DataFrame(data)

# Print the DataFrame
display(df)
```
